## Binomial distribution explaination

I roll a dice 12 times. What is the probability of getting a 4 "X" no. of times. 
We get a list of probabilities with 12 times. That is probability of getting a 4 0 times, 1 times, 2 times.........12 times.

Mean of binomial distribution = np
Variance of binomial distribution = np(1-p)

```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 12
p = 1/6
bd_list = []
n_list = np.arange(0,n+1)
for i in range(0,n+1):
    r = comb(n,i)*(p**i)*((1-p)**(n-i))
    bd_list.append(r)
    
```

## Normal Distribution

Normal distribution is just like binomial distribution or poisson distribution withh its own parameter and it is a very good approximation for binomial distribution

```python
v = n*p*(1-p)
v
sd = math.sqrt(v)
mean = n*p
mean
v
sd
```

```python
nd_list = []
for i in range(0,n+1):
    k = math.e**((-0.5)*((i-mean)/sd)**2)
    r = k/(sd*math.sqrt(2*math.pi))
    nd_list.append(r)
```

```python
# This plot shows normal distribution
plt.plot(n_list, nd_list)
```

```python
# This plot shows both normal distribution (in red) and binomial distribution (in blue)
plt.figure(figsize=(15,5))
plt.plot(n_list, nd_list, 'r')
plt.plot(n_list, bd_list, 'b')
```
