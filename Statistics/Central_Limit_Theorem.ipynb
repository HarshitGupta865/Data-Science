The Central Limit Theorem states that, given a large enough sample size from a population with any shape of distribution, the sampling distribution of the mean will be approximately normal, or bell-shaped, regardless of the shape of the original population. This means that, as the sample size increases, the mean of the samples will get closer and closer to the true mean of the population.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Variables and their usage

1. N = Number of elements in the population
2. pop_arr = Array of elements in population
3. rep = No. of times sample is taken from population
4. sample_size = No. of elements in each sample
5. sample_arr = Elements in each sample
6. sample_arr_mean = Array of mean of each array of sample_arr

```python
N = 1000
pop_arr = np.random.randint(0, 100, size=N)
print(pop_arr)
```

```python
plt.hist(pop_arr)
```

```python
print("Population mean = ", np.mean(pop_arr))
print("Population Std = ", np.std(pop_arr))
```

```python
rep = 5000
sample_size = 25
sample_arr = []

for i in range(0,rep):
    sample_arr.append(np.random.choice(pop_arr,5))

sample_arr
```

```python
sample_arr_mean = []

for i in sample_arr:
    sample_arr_mean.append(i.mean())
```

```python
sample_arr_mean
```

```python
sns.distplot(sample_arr_mean, bins=10)
```

```python
print("Population mean = ", np.mean(pop_arr))
print("Population Std = ", np.std(pop_arr))
print("----------------------------------------------------------")
print("Sample mean = ", np.mean(sample_arr_mean))
print("Sample Std = ", np.std(sample_arr_mean))
```

Above conclusion shows that Sample mean is very closely equal to population mean
And sample standard deviation = population SD/sq root of size of each sample
