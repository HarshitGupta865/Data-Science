# Matplotlib Part - 1

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD OF USING MATPLOTLIB
plt.plot(x,y)
```

```python
# To add x-label or y-label or title of the plot
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Creating multi plot on the same canvas
# Creating a plot on 1st row, 2nd column, 1st picture
plt.subplot(1,2,1)
plt.plot(x,y,'r')

# Creating a plot on 1st row, 2nd column, 2nd picture
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OBJECT ORIENTED METHOD OF USING MATPLOTLIB

# Consider fig as just a imaginary blank canvas
fig = plt.figure()

# Add axes to above canvas
# Takes list of 4 parameter(left, bottom, width, height)
axes = fig.add_axes([0.1,0.1,0.8,0.8])

# Plotting on axes
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title("Title")
```

```python
# Putting multiplots on one canvas using object oriented method

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

# Matplotlib Part - 2

```python
# Creating subplots using object oriented method

fig,axes = plt.subplots(nrows=1, ncols=2)

# Here axes is just a list of 2 axes object

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

### Figure Size, aspect ratio & DPI

```python
fig = plt.figure(figsize=(3,2), dpi=1000)

ax =  fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,3))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
# To save a figure
fig.savefig('my_pic.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

# ax.legend(loc=10)
ax.legend(loc=0)
```
