# Distribution Plots

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loadiing a built in dataset in seaborn called tips
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Dist Plot - Allows us to show the distribution of a univariate set of observations
# Univaraite means one variable


# sns.distplot(tips['total_bill'])
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
# Joint Plot - Allows you to match up two dist plots for bivariate data
# Bivariate means two variables


sns.jointplot(x='total_bill', y='tip', data=tips)
# sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
# sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
# sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
# Pair plot - Allows you to plot pairwise relationships across entire dataframe atleast for numerical columns
# sns.pairplot(tips)
# sns.pairplot(tips, hue='sex')

sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
# Rug Plot - to explain construction of kde along plots
sns.rugplot(tips['total_bill'])
```

# Categorical Plots

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
# Count plot - Simply counts the no.of occurence of categories in a column
sns.countplot(x='sex', data=tips)
```

```python
# Box Plot
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# Violin plot
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
# Strip Plot
# sns.stripplot(x='day', y='total_bill', data=tips)
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
# Swarm Plot - Combination of strip plot and violin plot
# sns.swarmplot(x='day', y='total_bill', data=tips)

sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```
