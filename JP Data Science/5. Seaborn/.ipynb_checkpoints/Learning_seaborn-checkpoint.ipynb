# Distribution Plots

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# sns.distplot(tips['total_bill'])
# sns.distplot(tips['total_bill'], kde=False)
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
#Syntax for distplot ------>  sns.jointplot(x=, y=, data=)

sns.jointplot(x='total_bill', y='tip', data=tips)
# sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
# sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
# sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
# sns.pairplot(tips)
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical Plots

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker', jitter=True)
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
#Heatmap
tc = tips.corr()
# sns.heatmap(tc)
# sns.heatmap(tc, annot=True)
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=1)
```

```python
sns.clustermap(fp)
```

# Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
g = sns.PairGrid(iris)
# g.map(plt.scatter)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
# sns.lmplot(x='total_bill', y='tip', data=tips)
# sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
# sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['v', 'o'])
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['v', 'o'], scatter_kws={'s':100})
```

```python
# sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
# sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
# sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='time', hue='sex')
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex')
```

# Style & Color

```python
# sns.set_style(style='darkgrid')
sns.set_style(style='whitegrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style(style='ticks')
sns.countplot(x='sex', data=tips)
# sns.despine()
sns.despine(left=True, bottom=True)
```

```python
sns.set_style('ticks')
sns.set_context('notebook')
sns.countplot(x='sex', data=tips, palette='seismic')
```
