## Correcting the format of each column in the dataset

```python
df.info()
```

We need to convert Reviews column from object datatype to integer data type


We need to convert size column from object datatype to float data type after correcting the format of this column.


We need to convert install column from object datatype to integer data type after correcting the format of this column.


We need to convert price column from object datatype to integer data type after correcting the format of this column.

```python
# Converting the data type of reviews column
df["Reviews"] = df["Reviews"].astype('int64')
```

```python
# we want to Replace all the unwanted character to convert datatye to numeric 
def clean_it(num):
  """This function takes a string and replace the following characters if present, '+', ',' ,'$', 'M', 'k', 'NaN'"""
  if '+' in num:
    num = num.replace('+','')
  if ',' in num:
    num = num.replace(',','')
  if '$' in num:
    num = num.replace('$','')
  if 'M' in num:
    num = str(int(float(num.replace('M',''))*1000))
  if 'k' in num:
    num = str(int(float(num.replace('k',''))))
  if 'NaN' in num:
    num = '0'
  else:
    pass
  return num
```

```python
#Changing the 'Reviews' column values into valid numeric values
df['Reviews'] = pd.to_numeric(df['Reviews'])

#Changing the 'Size' column values into valid numeric values
df['Size'] = df['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)
df['Size'] = pd.to_numeric(df['Size'].map(lambda x: clean_it(x)))

#Changing the 'Installs' column values into valid numeric values
df['Installs'] = pd.to_numeric(df['Installs'].map(lambda x: clean_it(x)))

#Changing the 'Price' column values into valid numeric values
df['Price'] = pd.to_numeric (df['Price'].map(lambda x:clean_it(x)))
df.info()
```

### 13. Sentiment Analysis Based on Category

```python
from matplotlib.ticker import PercentFormatter
merged_df = pd.merge(df, ur_df, on='App', how = 'inner')
merged_df.head()
f = plt.figure(figsize=(15,8))
ax = f.add_subplot(1,1,1)
#sns.set_theme(style="ticks")

sns.histplot(
    data=merged_df,
    x="Category", hue="Sentiment",
    bins=34,
    ax=ax, 
    stat="count",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
    legend=True
    )


ax.set_title("Sentiment Analysis Based on Category",fontsize=15,fontweight='bold')
plt.xticks(rotation='vertical')
ax.set_xlabel("Category",fontsize=14)
ax.set_ylabel("Review Counts",fontsize=14)

plt.gca().yaxis.set_major_formatter(PercentFormatter(20000))
sns.set(style="ticks")
plt.grid()
plt.show()
```

Inference:

Here we notice that most of the reaction are from 'Game' catagory and very less from 'Comics', 'Events', 'Maps_And_Navigation' and 'Weather'.
