In linear regression we have a straight line f(x) also called regression line and this line predicts how good is the fit.
Now how do we get f(x) such that it is closest to the actual data in the dataset. That means how do we get an f(x) for which cost function is minimum.

Lets check how cost function vary with f(x)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480,  430,   630, 730,])
```

```python
x_mean = np.mean(x_train)
y_mean = np.mean(y_train)
prod = x_mean*y_mean
prod_train = x_train*y_train
prod_mean = np.mean(prod_train)
x_mean_sq = x_mean**2
x_square = x_train*x_train
x_square_mean = np.mean(x_square)
```

```python
plt.plot(x_train,y_train, marker='x')
```

Lets take slope = 100 and y-intercept = 100. That means we are taking reg line **f(x) = 100x+100** . Let us create a line with this f(x)

```python
def calc_reg_line(m, b, x):
    
    #Calculating the no. of training samples in n
    n = x.shape[0]
    
    #Creating an array of zeroes equal to n in f
    f = np.zeros(n)
    
    for i in range(n):
        f[i] = m*x[i] + b
    
    return f
        
```

```python
m = (prod - prod_mean)/(x_mean_sq - x_square_mean)
b = y_mean - m*x_mean
res_f = calc_reg_line(m, b, x_train)
print(res_f)
```

```python
# Below visualisation to show the actual data points and the straight line
plt.scatter(x_train, y_train, marker='x', c='r')
plt.plot(x_train, res_f, c='green')
```

Lets see the cost function graph for **f(x) = 100x + 100**

```python
def calc_cost_func(f, y):
    
    j = np.zeros(n)
    min_sum = 0
    
    for i in range(1,n+1):
        j[i-1] = ((f[i-1]-y[i-1])**2)/(2*n)
        min_sum = min_sum + j[i-1]
        
    print(min_sum)
        
    return j
```

```python
res_j = calc_cost_func(res_f, y_train)
print(res_j)
```
