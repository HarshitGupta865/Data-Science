Let us have 2 dataset in training set and 2 in target dataset

y = 300, 500

```python
import numpy as np
import matplotlib.pyplot as plt
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

print (f'x_train = {x_train}')
print (f'y_train = {y_train}')
```

```python
# Creating a visualisation for real dataset
plt.scatter(x=x_train, y=y_train, marker='x', c = 'r')
plt.xlabel('Size of land')
plt.ylabel('Price')
plt.title('Housing Price')
```

```python
slope = 100
b = 100
```

```python
# Now, lets create a logic to define the straight line ie, the regression line

def calc_reg_line(x,slope,b ):
    
    m = x_train.shape[0]
    f = np.zeros(m)
    
    for i in range(m):
        f[i] = slope*x[i] + b
    
    return f
    
```

```python
temp_f = calc_reg_line(x_train, slope, b)
print(temp_f)
```

```python
plt.plot(x_train, temp_f, c='green', label="Prediction")
plt.scatter(x_train, y_train, c='red', label="Actual Data")


# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

As we see above we need to calculate the function 'f' so that we can get as greater a fit for the data.
